## Cities for Season Lovers

```python
allData = [chicagoData] + [houstonData] + [miamiData] + [newYorkData] + [sanFranciscoData] + [seattleData]
cities = ["Chicago", "Houston", "Miami", "NewYork", "SanFrancisco", "Seattle"]
maxes = [max(data) - min(data) for data in allData]
x = np.arange(1,7,1)

cities = [i for _, i in sorted(zip(maxes, cities))]
maxes.sort()

ax1 = plt.subplot()
ax1.set_xticks(x)
ax1.set_xticklabels(cities)

plt.bar(x , maxes)
```

## Sunniest Summers

```python

cities = ["Chicago", "Houston", "Miami", "NewYork", "SanFrancisco", "Seattle"]
summerData = [100*sum(i[5:8])/((30 + 31 + 31)*24) for i in allData]
print(summerData)

x = np.arange(1,7,1)

cities = [i for _, i in sorted(zip(summerData, cities))]
summerData.sort()

ax1 = plt.subplot()
ax1.set_xticks(x)
ax1.set_xticklabels(cities)

plt.bar(x , summerData)
plt.ylim(25,45)
```
